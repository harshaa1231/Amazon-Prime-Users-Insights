```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
amazon['Feedback/Ratings'].plot(kind='hist', bins=20, title='Feedback/Ratings')

# Removing top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel('Feedback/Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=amazon, x='Gender')
```

```python
payment_info_counts = amazon.groupby('Payment Information').size()
sns.set(style="whitegrid")
payment_info_counts.plot(kind='barh', color=sns.color_palette('Dark2'))
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.title('Distribution of Payment Information')
plt.xlabel('Count')
plt.ylabel('Payment Information')
plt.show()
```

```python
counts = amazon['Purchase History'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=100)
plt.title('Pie Chart of ' + 'Purchase History')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()  # Display the pie chart
plt.show()
```

```python
Devices_Used = amazon['Devices Used'].value_counts() # Get counts for each device type

plt.figure(figsize=(8, 8))  # Adjust figure size as needed
plt.hist(amazon['Devices Used'], color=sns.color_palette('Dark2')[1]) # Use Matplotlib's hist function
plt.title('Distribution of Devices Used by Subscribers')
plt.xlabel('Device Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.hist(amazon['Renewal Status'], bins= 7, color= 'Green')
plt.xlabel('Renewal Status')
plt.ylabel('Count of the people')
plt.title('Histogram of Renewal status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
amazon['Purchase History'].value_counts().plot(kind='bar', color='Black')
plt.xlabel('Purchase History')
plt.ylabel('Count')
plt.title('Count of Purchase History')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
amazon['Favorite Genres'].value_counts().plot(kind='bar', color='Maroon')
plt.xlabel('Favorite Genre')
plt.ylabel('Count')
plt.title('Count of Favorite Genre')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.lineplot(x='Usage Frequency', y='count', data=amazon.groupby('Usage Frequency').size().reset_index(name='count'))
plt.title('Line Chart of Usage Frequency')
plt.xlabel('Usage Frequency')
plt.ylabel('Count')
plt.show()
```

```python
heatmap = amazon.groupby('Renewal Status')['Usage Frequency'].value_counts().unstack().fillna(0)
plt.figure(figsize=(8, 8))
sns.heatmap(heatmap, cmap='viridis', annot=True, fmt='.0f')  # annot=True to show values, fmt='.0f' for integer format
plt.xlabel('Usage Frequency')
plt.ylabel('Renewal Status')
plt.title('Heatmap of Renewal Status vs Usage Frequency')
plt.show()
```

```python
grouped_data = amazon.groupby(['Gender', 'Favorite Genres']).size().unstack()
bar_of_Comparsion = grouped_data.plot(kind='bar', figsize=(10, 6))
bar_of_Comparsion.set_xlabel('Favorite Genres')
bar_of_Comparsion.set_ylabel('Count')
bar_of_Comparsion.set_title('Comparison of Favorite Genres by Gender')
bar_of_Comparsion.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
engagement_metrics = pd.crosstab(amazon['Devices Used'], amazon['Engagement Metrics'])
engagement_metrics_percent = engagement_metrics.div(engagement_metrics.sum(axis=1), axis=0)
engagement_metrics_percent.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Devices Used')
plt.ylabel('Percentage of Subscribers')
plt.title('Comparison of Engagement Metrics by Devices Used')
plt.legend(title='Engagement Metrics')
plt.xticks(rotation=45)
plt.show()
```

```python
grouped_data = amazon.groupby(['Gender', 'Purchase History']).size().unstack()
gender_vs_ph = grouped_data.plot(kind='bar', figsize=(10, 6))
gender_vs_ph.set_xlabel('Purchase History')
gender_vs_ph.set_ylabel('Count')
gender_vs_ph.set_title('Purchase History by Gender')
gender_vs_ph.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x="Engagement Metrics", y="Feedback/Ratings", data=amazon)
plt.xlabel("Engagement Metrics")
plt.ylabel("Feedback/Ratings")
plt.title("Engagement Metrics vs Feedback/Ratings")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract year and month from Membership Start Date and Membership End Date
amazon['Membership Start Year'] = pd.to_datetime(amazon['Membership Start Date']).dt.year
amazon['Membership Start Month'] = pd.to_datetime(amazon['Membership Start Date']).dt.month
amazon['Membership End Year'] = pd.to_datetime(amazon['Membership End Date']).dt.year
amazon['Membership End Month'] = pd.to_datetime(amazon['Membership End Date']).dt.month

# Create two subplots
fig, (axis1, axis2) = plt.subplots(1, 2, figsize=(18, 6))

# Bar chart for Membership Start Year
axis1.bar(amazon['Membership Start Year'].value_counts().index, amazon['Membership Start Year'].value_counts().values)
axis1.set_xlabel('Year')
axis1.set_ylabel('Count')
axis1.set_title('Membership Start Year')

# Bar chart for Membership End Year
axis2.bar(amazon['Membership End Year'].value_counts().index, amazon['Membership End Year'].value_counts().values)
axis2.set_xlabel('Year')
axis2.set_ylabel('Count')
axis2.set_title('Membership End Year')
plt.show()
```

```python
# Create a cross-tabulation of Usage Frequency and Devices Used
usage_device_crosstab = pd.crosstab(amazon['Usage Frequency'], amazon['Devices Used'])

# Create a heatmap
sns.heatmap(usage_device_crosstab, cmap='viridis', annot=True, fmt='.0f')

# Add labels and title
plt.xlabel('Devices Used')
plt.ylabel('Usage Frequency')
plt.title('Usage Frequency vs. Devices Used')

# Show the plot
plt.show()
```
